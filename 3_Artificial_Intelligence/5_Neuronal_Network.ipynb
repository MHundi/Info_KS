## Neuronale Netzwerke

### Universal approximation theorem:




Neuronale Netze mit nur einem Hidden Layer, welcher beliebig viele Neuronen aufweisen kann, können jede stetige Funktion nachbilden, wenn die verwendeten Aktivierungsfunktionen differenzierbar sind und für $x\to \pm \infty$
die Funktionswerte zwischen 0 und 1 liegen. 

Heavisidefunktion
$$f(x)=\begin{cases}
0 & x\leq 0\\
1 & x \geq 0 
\end{cases}
$$

```python
x=np.linspace(-5,5,100)
y=np.heaviside(x,1)
plt.plot(x,y)
```

Logistische Funktion (S-Funktion)

$$
f(x)=\frac{1}{1+e^{-x}}
$$

```python
x=np.linspace(-5,5,100)
y=1/(1+np.exp(-x))
plt.plot(x,y)
```

Eigenschaften der Logistischen Funktion:

- $\lim_{x\to - \infty} = 0$

- $\lim_{x\to 0 \infty} = 1$

- Für $x \in \mathbb{R}$ differenzierbar 
    

Bemerkung: Gauß-Funktion
$$
f(x)=\frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left( \frac{x-\mu}{\sigma} \right)^2}
$$

Für $\sigma =1$ und $\mu =0$ gilt:

```python
def gaussian(x, mu, sig):
    return (
        1.0 / (np.sqrt(2.0 * np.pi) * sig) * np.exp(-np.power((x - mu) / sig, 2.0) / 2)
    )
x=np.linspace(-5,5,100)
y=gaussian(x,0,1)
plt.plot(x,y)
```

Die Ableitung der Gauß-Funktion für $\sigma = 1$ und $\mu = 0$:
$$f'(x)=\frac{1}{2\pi}(-x)\cdot e^{-\frac{1}{2}x^2}
$$

```python
def dx_gaussian(x, mu, sig):
    return (
        1.0 / 2*np.pi*(-x)* np.exp(-np.power(x,2))
    )
x=np.linspace(-5,5,100)
y=dx_gaussian(x,0,1)
plt.plot(x,y)
```
